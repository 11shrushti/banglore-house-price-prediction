```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
df1 = pd.read_csv("/content/Bengaluru_House_Data.csv")
df1.head()
```

```python
df1.shape
```

```python
df1.groupby("area_type")["area_type"].count()
```

```python
# doing to drop the columns which are not required

df1.drop(columns=["area_type","society","balcony","availability"],inplace = True)
df1.head()
```

```python
# find the nan values
df1.isnull().sum()
```

```python
# drop the nan values rows
df2 = df1.dropna()
df2.isnull().sum()
```

```python
# check indivisual itmes in the size column
df2["size"].unique()
```

```python
# make a new column and create a function

df2["bhk"] = df2["size"].apply(lambda x: int(x.split(" ")[0]))
```

```python
df2.head()
```

```python
df2["bhk"].unique()
```

```python
# check for the total sqt
df2.total_sqft.unique()
```

```python
# we need to convert the range into the avg and convert it to a singlr number
def is_float(x):
  try:
    float(x)
  except:
    return False
  return True

df2[~df2["total_sqft"].apply(is_float)].sample(20)  # make it oppoosite
```

```python
# we will do data cleaning and then we will find the avg of the range and then we will ignore the other values

def conver_sqft_to_num(x):
  token = x.split("-")
  if len(token)==2:
    return (float(token[0])+ float(token[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
df3 = df2.copy()
df3["total_sqft"] = df3["total_sqft"].apply(conver_sqft_to_num)
df3.sample(15)
```

```python
# make a copy of the dataframe and make changes for the price column

df4 = df3.copy()
df4["price_per_sqft"] = df4["price"]*100000/df4["total_sqft"]
df4.head()
```

```python
# check location we have in total
len(df4.location.unique())
```

```python
# find the number of locations

df4.location = df4.location.apply(lambda x: x.strip())

location_stats = df4.groupby("location")["location"].count().sort_values(ascending=False)


location_stats
```

```python
# we will condier the location with less than 10 count as other locaion

len(location_stats[location_stats<=10])
```

```python
location_stats_less_than_10 = location_stats[location_stats <=10]
location_stats_less_than_10
```

```python
df4.location = df4.location.apply(lambda x : "other" if x in location_stats_less_than_10 else x)
len(df4.location.unique())
```

```python
df4[df4.total_sqft/df4.bhk<300].head()
```

```python
# we are going to remove all the data points from there

df4.shape
```

```python
# we will remove all the outlier from the data

df5 = df4[~(df4.total_sqft / df4.bhk < 300)]
df5.shape
```

```python
# we will check if the column pricr_per_sqft have any outhliers

df5.price_per_sqft.describe()
```

```python
# we will find per location the mean and the standard deviation and we will remove the extreme case

def remove_pps_outliers(df):
  df_out = pd.DataFrame()
  for key , subdf in df.groupby("location"):
    m = np.mean(subdf.price_per_sqft)
    st = np.std(subdf.price_per_sqft)
    reduced_df = subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft<=(m+st))]
    df_out = pd.concat([df_out , reduced_df],ignore_index =True)
  return df_out


df6 =  remove_pps_outliers(df5)
df6.shape
```

```python
# we will create a visualization where we check if the 2bhk flats are higher then the 3bhk falts

def plot_scatter_chart(df,location):
  bhk_2 = df[(df.location == location) & (df.bhk == 2)]
  bhk_3 = df[(df.location == location) & (df.bhk == 3)]
  plt.scatter(bhk_2.total_sqft , bhk_2.price, color = "blue" , label = "2 BHK", s=50)
  plt.scatter(bhk_3.total_sqft , bhk_3.price , marker="+", color = "green", label ="3 BHK",s=50)
  plt.xlabel("Total Square Feet Area")
  plt.ylabel("Price Per Square Feet")
  plt.title(location)
  plt.legend()

plot_scatter_chart(df6,"Hebbal")
```

```python
# we can find that the price of the 2bhk are high then 3 bhks we have to remoe this outlier

def remove_bhk_outlier(df):
  exclude_indices = np.array([])
  for location , location_df in df.groupby("location"):
    bhk_stats = {}
    for bhk , bhk_df in location_df.groupby("bhk"):
      bhk_stats[bhk] = {
          "mean" : np.mean(bhk_df.price_per_sqft) ,
          "std" : np.std(bhk_df.price_per_sqft),
          "count": bhk_df.shape[0]
          }
    for bhk , bhk_df in location_df.groupby("bhk"):
      stats = bhk_stats.get(bhk-1)
      if stats and stats["count"]>5:
        exclude_indices = np.append(exclude_indices , bhk_df[bhk_df.price_per_sqft<(stats["mean"])].index.values)
  return df.drop(exclude_indices , axis ="index")

df7 = remove_bhk_outlier(df6)
df7.shape
```

```python
df8 = df7[df7.bath < df7.bhk+2]
df8.shape
```

```python
# we will drop these two feaure the size and price per sqft

df9 = df8.drop(columns=["size","price_per_sqft"])
df9.sample(5)
```

```python
# we have to convert the string column to the integer

dummies = pd.get_dummies(df9.location)
dummies.sample(3)
```

```python
df10 = pd.concat([df9,dummies.drop("other",axis = "columns")],axis = "columns")# in dummies we drop the first columns
df10.sample(5)
```

```python
df11 = df10.drop("location",axis = "columns")
df11.sample(5)
```

```python
df11.shape
```

```python
X = df11.drop("price",axis = "columns")
X.head()
```

```python
Y= df11.price
Y.head()
```

```python
# we will divide the dataset into train and split

from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size = 0.2,random_state=10)
```

```python
# we will do the linear regreesion

from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train , Y_train)
lr_clf.score(X_test, Y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits = 5 , test_size = 0.2 , random_state = 0)

cross_val_score(LinearRegression() , X ,Y,cv =cv)
```

```python
from sklearn.model_selection import GridSearchCV # here we tru out the differnet models and check which one suits the more


from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
  algos = {
      "linear_regression" :{
          "model" :LinearRegression(),
          "params" : {
              # "normalize" : [True, False]
      }
  },
  "lasso":{
    "model": Lasso(),
    "params" : {
      "alpha" : [1,2],
      "selection" : ["random","cyclic"]
  }
  },
      "decision_tree" : {
          "model" : DecisionTreeRegressor(),
          "params" : {
              "criterion" : ["mse","friedman_mse"],
              "splitter" : ["best","random"]
          }
  }
  }
  scores = []
  cv = ShuffleSplit(n_splits = 5 , test_size = 0.2 , random_state = 0)
  for algo_name , config in algos.items():
    gs = GridSearchCV(config["model"], config["params"], cv=cv, return_train_score=False)
    gs.fit(X,Y)
    scores.append({
        "model" : algo_name,
        "best_score" : gs.best_score_,
        "best_params" : gs.best_params_
    })
  return pd.DataFrame(scores, columns=["model","best_score","best_params"])

find_best_model_using_gridsearchcv(X,Y)



```

```python
def predict_price(location,sqft,bath,bhk):
  loc_index = np.where(X.columns == location)[0][0]

  x = np.zeros(len(X.columns))
  x[0] = sqft
  x[1] = bath
  x[2] = bhk
  if loc_index >= 0:
    x[loc_index] = 1

  return lr_clf.predict([x])[0]
```
